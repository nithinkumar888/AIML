```python
#part01
import numpy as np

# Creating arrays from lists
lst1 = [1, 2, 3]
array1 = np.array(lst1)

list2 = [3, 2, 1]
array2 = np.array(list2)

# Operations on arrays
print("Array 1:", array1)
print("Array 2:", array2)

#  multiplication, division, and exponentiation
print("Array2 multiplied by Array1: ", array1 * array2)
print("Array2 divided by Array1: ", array2 / array1)
print("Array2 raised to the power of Array1: ", array2 ** array1)

# Mathematical functions
print("Sine: ", np.sin(array1))
print("Natural logarithm: ", np.log(array1))
print("Base-10 logarithm: ", np.log10(array1))
print("Base-2 logarithm: ", np.log2(array1))
print("Exponential: ", np.exp(array1))

# Adding arrays
print(f"Adding two numpy arrays {array1} and {array2} together: {array1 + array2}")


#part02

import matplotlib.pyplot as plt

# Data for visualization
people = ['Ann', 'Brandon', 'Chen', 'David', 'Emily', 'Farook',
          'Gagan', 'Hamish', 'Imran', 'Julio', 'Katherine', 'Lily']
age = [21, 12, 32, 45, 37, 18, 28, 52, 5, 40, 48, 15]
weight = [55, 35, 77, 68, 70, 60, 72, 69, 18, 65, 82, 48]
height = [160, 135, 170, 165, 173, 168, 175, 159, 105, 171, 155, 158]

# Scatter plot: Relationship between Weight and Height
plt.scatter(weight, height)
plt.title("Relationship between Weight and Height of Patients")
plt.ylabel("Height (cm)")
plt.xlabel("Weight (kg)")
plt.show()

# Bar plot: People's weight in kgs
plt.figure(figsize=(12, 4))
plt.title("People's weight in kgs", fontsize=16, fontstyle='italic')
plt.bar(x=people, height=weight, width=0.6, color='orange', edgecolor='k', alpha=0.6)
plt.xlabel("People", fontsize=15)
plt.ylabel("Weight (in kgs)", fontsize=15)
plt.show()

# Histogram: Distribution of weight
plt.figure(figsize=(7, 5))
plt.hist(weight, color='red', edgecolor='k', alpha=0.75, bins=5)
plt.title("Histogram of patient weight", fontsize=18)
plt.xlabel("Weight in kgs", fontsize=15)
plt.show()


#part03

import numpy as np
import pandas as pd

# Reading data from text file (tab-delimited)
df1 = pd.read_table("Lab01-Text.txt")
print(df1)

# Reading data from CSV file
data2 = pd.read_csv("Lab01-Text.csv")
print(data2)

# Reading data from Excel file
# Ensure you have 'openpyxl' installed to read .xlsx files
data3 = pd.read_excel("Height_weight.xlsx")
print(data3)
```
